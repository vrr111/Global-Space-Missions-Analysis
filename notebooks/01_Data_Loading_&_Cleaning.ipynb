```python
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
```

```python
# Calculate company-wise mean price
company_mean = df.groupby("Company")["Price"].transform("mean")

# Fill missing prices
df["Price"] = df["Price"].fillna(company_mean)

# Fill any remaining missing values with the overall mean
df["Price"] = df["Price"].fillna(df["Price"].mean())
```

```python
# Check missing values
print(df["Time"].isnull().sum())

# Remove rows where Time is missing
df = df.dropna(subset=["Time"])

# Reset index
df.reset_index(drop=True, inplace=True)

# Verify
print(df["Time"].isnull().sum())
print(df.shape)
```

```python
# Convert Time from HH.MM.SS to HH:MM:SS

df["Time"] = (
    df["Time"]
    .astype(str)
    .str.replace(".", ":", regex=False)
)
df.head()
```
