```python
# Import libraries

import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
# Load cleaned dataset

df = pd.read_csv(
    
    "outputs/cleaned_dataset.csv",
    parse_dates=["Date"]
)

# Create Launch_DateTime column

df["Launch_DateTime"] = pd.to_datetime(
    df["Date"] + " " + df["Time"].str.replace(".", ":", regex=False),
    format="%d-%m-%Y %H:%M:%S"
)

df[["Date", "Time", "Launch_DateTime"]].head()
```

```python
# Extract launch year
df["Launch_Year"] = df["Launch_DateTime"].dt.year

# Extract launch month
df["Launch_Month"] = df["Launch_DateTime"].dt.month

# Extract month name
df["Launch_Month_Name"] = df["Launch_DateTime"].dt.month_name()

# Extract quarter
df["Launch_Quarter"] = df["Launch_DateTime"].dt.quarter

# Extract day of month
df["Launch_Day"] = df["Launch_DateTime"].dt.day

# Extract weekday name
df["Launch_Day_Name"] = df["Launch_DateTime"].dt.day_name()

# Extract ISO week number
df["Launch_Week"] = df["Launch_DateTime"].dt.isocalendar().week

# Create launch decade - FIXED: Extract year first, then calculate decade
df["Launch_Decade"] = (df["Launch_DateTime"].dt.year // 10) * 10

# Extract launch hour
df["Launch_Hour"] = df["Launch_DateTime"].dt.hour

# Extract launch minute
df["Launch_Minute"] = df["Launch_DateTime"].dt.minute

# Extract launch second
df["Launch_Second"] = df["Launch_DateTime"].dt.second
```

```python
# Categorize launch period

def launch_period(hour):

    if hour < 6:
        return "Night"

    elif hour < 12:
        return "Morning"

    elif hour < 18:
        return "Afternoon"

    else:
        return "Evening"

df["Launch_Period"] = df["Launch_Hour"].apply(launch_period)
```

```python
# Extract country from location

df["Country"] = df["Location"].str.split(",").str[-1].str.strip()

# Extract launch site

df["Launch_Site"] = df["Location"].str.split(",").str[0].str.strip()
```

```python
# Categorize rocket price

df["Price_Category"] = pd.cut(
    df["Price"],
    bins=[0,25,50,75,100,np.inf],
    labels=[
        "Low",
        "Medium",
        "High",
        "Very High",
        "Premium"
    ]
)
```

```python
# Simplify mission result

df["Mission_Result"] = np.where(
    df["MissionStatus"] == "Success",
    "Successful",
    "Unsuccessful"
)
```

```python
# Copy rocket status

df["Rocket_Availability"] = df["RocketStatus"]
```

```python
# Save feature engineered dataset

df.to_csv(
    "outputs/space_missions_feature_engineered.csv",
    index=False
)

print("Feature engineered dataset saved successfully.")
```
