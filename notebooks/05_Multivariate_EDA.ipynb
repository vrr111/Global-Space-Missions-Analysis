```python
# Import libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

```python
# Load feature engineered dataset

df = pd.read_csv("outputs/space_missions_feature_engineered.csv")

df.head()
```

```python
# Convert Launch_DateTime to datetime

df["Launch_DateTime"] = pd.to_datetime(df["Launch_DateTime"])
```

```python
# Check dataset information

df.info()
```

```python
# Company, Country and Mission Status

company_country = pd.pivot_table(
    df,
    index="Company",
    columns="Country",
    values="Mission_Result",
    aggfunc="count",
    fill_value=0
)

company_country.head()
```

```python
# Company, Rocket and Mission Status

company_rocket = pd.crosstab(
    [df["Company"], df["Rocket"]],
    df["MissionStatus"]
)

company_rocket.head()
```

```python
# Country, Launch Year and Mission Status

country_year = pd.pivot_table(
    df,
    index="Launch_Year",
    columns="Country",
    values="Mission_Result",
    aggfunc="count",
    fill_value=0
)

country_year.head()
```

```python
# Launch Period, Rocket Status and Mission Status

period_status = pd.crosstab(
    [df["Launch_Period"], df["RocketStatus"]],
    df["MissionStatus"]
)

period_status
```

```python
# Average launch price by company

company_price = df.groupby("Company").agg(
    Average_Price=("Price", "mean"),
    Successful_Missions=("Mission_Result", "count")
)

company_price.head()
```

```python
# Country-wise launch statistics

country_price = df.groupby("Country").agg(
    Average_Price=("Price", "mean"),
    Total_Missions=("Mission_Result", "count")
)

country_price
```

```python
# Company launches by year

company_year = pd.crosstab(
    df["Launch_Year"],
    df["Company"]
)

company_year.head()
```

```python
# Launches by year and country

year_country = pd.pivot_table(
    df,
    index="Launch_Year",
    columns="Country",
    values="Mission_Result",
    aggfunc="count",
    fill_value=0
)

year_country.head()
```
